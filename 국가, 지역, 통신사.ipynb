### user_ip를 이용하여 국가, 지역, 통신사 알아내기

```python
# 나라명, 지역명
import geoip2.database
from tqdm import tqdm

reader1 = geoip2.database.Reader('GeoLite2-City.mmdb')

countries = []
cities = []

for ip in tqdm(ive_join_info['user_ip']):
    if pd.isna(ip):
        countries.append(None)
        cities.append(None)

    else:
        try:
            response = reader1.city(str(ip))
            countries.append(response.country.iso_code)
            cities.append(response.city.name)
        except Exception:
            countries.append(None)
            cities.append(None)

ive_join_info['country'] = countries
ive_join_info['city'] = cities
```

```python
# 통신사
reader = geoip2.database.Reader("GeoLite2-ASN.mmdb")

carriers = []

for ip in tqdm(ive_join_info['user_ip']):
    if pd.isna(ip):
        carriers.append(None)
    
    else:
        try:
            response = reader.asn(str(ip))
            carriers.append(response.autonomous_system_organization)
        except Exception:
            carriers.append(None)

ive_join_info['carrier_1'] = carriers
```
